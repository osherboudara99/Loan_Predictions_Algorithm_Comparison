# Data Preprocessing

In order to build a sufficient ML model, we must change our categorical data into numeric data. Here, we will be using a data preprocessing method which will transform all the categorical data into numeric data. The code below was taken from 

```python
from sklearn.preprocessing import LabelEncoder

column_labels = ["Profession","CITY", "STATE","Married/Single", "House_Ownership", "Car_Ownership"]
train_data_categorical = train_data.copy()
test_data_categorical = test_data.copy()
```

```python
for i in column_labels:
    if(train_data_categorical[i].dtype=='object' and test_data_categorical[i].dtype=='object'):
        
        train_data_categorical[i].fillna('N', inplace=True)
        test_data_categorical.fillna('N', inplace=True)
        
        label = LabelEncoder()
        label.fit(train_data_categorical[i].values)
        train_data_categorical[i] = label.transform(train_data_categorical[i].values)
        
        label.fit(test_data_categorical[i].values)
        test_data_categorical[i]= label.transform(test_data_categorical[i].values)
```
